### Question 1 -- Word Composition Problem
- Write a program that:

1. Reads provided files (Input_01.txt) containing alphabetically sorted words list (one
word per line, no spaces, all lower case)
2. Identifies & display below given data in logs/console/output file

-- Longest compounded word

-- Second longest compounded word

-- Time taken to process the input file

Note: A compounded word is one that can be constructed by combining (concatenating) shorter words also found in the same file

```python
# Read the text file as mode r. 
file = open("C:/Users/prate/Downloads/Input_01.txt", "r")

# Create a list 
lists = []

# create a for loop to iterate through each line inthe given text file
for line in file:
    #line.strip() basically remove the line break in file
    stripped_line = line.strip()
    #stripped the line and create a list of word in the given line
    llist = stripped_line.split()
    #Append the previous list and try to add it with null list of the list
    lists.append(llist)

#close the file 
file.close()

#print the lists
print("Total Words in the list:")
print(lists)

#Define the node and initialize it
class Node:
    def __init__(self, alphabet=None, start=False):
        self.alphabet = alphabet
        self.start = start
        self.child = {}

# Create the trie and initialize & count it.
class Trie:
    def __init__(self):
        self.counts = {}
        self.root = Node()

#Insertion of the trie.
    def insert(self, word):
        current = self.root
        for alphabet in word:
            if alphabet not in current.child:
                current.child[alphabet] = Node(alphabet)
            current = current.child[alphabet]
        current.start = True

# Check alphabet are found in current child or not.
    def __contains__(self, word):
        current = self.root
        for alphabet in word:
            if alphabet not in current.child:
                return False
            current = current.child[alphabet]
        return current.start

# Return the total words that can make a group of word and subwords that can create a large word.
    def search(self, word):
        # It didn't take as empty string.
        if not word:
            # It return the 0 as the empty lists
            return 0, []
        # Try to search this word is found before or not and count it. 
        if word in self.counts:
            # if found then return and count the words.
            return self.counts[word]
        # If words are not found then we can do traversing.
        current = self.root
        # enumerate function allows the run iteration in for loops in for index and alphabet
        for index, alphabet in enumerate(word): 
            if alphabet not in current.child:
                return 0, []
            current = current.child[alphabet]
            if current.start:
                # Check suffix of word and Word + 1.
                suffix = word[index + 1:]    
                # Search the suffix list and suffix count
                suffix_count, suffix_list = self.search(suffix) 
                # Count the suffix and suffix_list
                self.counts[suffix] = suffix_count, suffix_list     
                # Check suffix count or not.
                if suffix_count:                                   
                    # Add prefix to the suffix list, add 1 to suffix count and return
                    return 1 + suffix_count, [word[:index + 1]] + suffix_list 
        # Return word that define in terminals
        return current.start, [word]
    
    # Returns the isCompound and list of small word
    def component(self, word):
        search_num, search_list = self.search(word)
        return search_num > 1, search_num, search_list

# Define the filename
def load(filename):
    words = []
    trie = Trie()
    # Opening the filename using with statement.
    with open(filename) as f:
        for line in f:
            # Line Strip basically remove the line break in the file.
            word = line.strip()
            # Insert word using trie
            trie.insert(word)
            # Append the word in given file 
            words.append(word)
    return trie, words

# Check the compound word in the text file
def pList(compoundList):
    #Sort the word in the compound list using lambda and when got it then reverse is equal to true.
    compoundList.sort(key = lambda tup: len(tup[0]), reverse=True)
    return compoundList

# Check the get compound word using trie and define compound is empty string
def componentList(trie, words):
    compound = []
    # In for loop it checks the conditions and return it.
    for word in words:
        isValid, num, dlist = trie.component(word)
        if isValid:
            # Append the word if valid
            compound.append((word, num, dlist))
    return compound

# Define the function and read the text file.
def example(filename='C:/Users/prate/Downloads/Input_01.txt'):
    # Try to load the filename
    trie, words = load(filename)
    # Check the compoundlist that define earlier in plist.
    compoundList = pList(componentList(trie, words))
    # Find the longestword and Secondlongest word using compoundlist
    longestWord = compoundList[0][0]
    secondLongestWord = compoundList[1][0]

    # Find the longest word and print it. 
    if len(longestWord): 
        print ("Longest compound word:\t\"" + longestWord + "\"")
    else:
        print ("Not found")
        
    # find the Secondlongest word and print it.
    if len(secondLongestWord):
        print ("Second Longest compound word:\t\"" + secondLongestWord + "\"")
    else:
        print ("Not found")
    return trie, words, compoundList

# Main function and read the file.
def main():
    filename = ("C:/Users/prate/Downloads/Input_01.txt")
    example(filename)

main()
```
